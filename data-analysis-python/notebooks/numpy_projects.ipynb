## Student Marks Analyzer

##### In this we create a 2-d np-array which represents students as row and subjects as columns

##### Perform actions to calculate
1. Average marks per student
2. Average marks per subject
3. Topper student index
4. Subject with highest average

```python
import numpy as np
```

```python
student_list = np.array([[85,90,93,74,78],[58,70,63,84,98],[95,80,73,84,98],[75,80,63,94,88],[95,70,83,94,68]])
student_list
```

```python
np.mean(student_list,axis=1)
```

```python
np.mean(student_list,axis = 0)
```

```python
np.argmax(np.sum(student_list,axis = 1))
```

```python
np.argmax(np.mean(student_list,axis=1))
```

```python
student_list.mean(axis=0)
```

### Project 3 - Image as A Matrix

```python
image = np.random.randint(0,256,(100,100))
image
```

```python
image[40:60,40:60]
```

```python
image + 50
```

```python
np.clip(image-50,0,255)
```

```python
image[image > 128] = 255
image
```

```python
image[image < 128] = 0
image
```

##### `np.where(arr,if_true,if_false)` is used instead of boolean indexing manually

##### Here we are increasing the value to maximum for the ones which is higher than 128

```python
np.where(image > 128,255,0)
```

##### To Invert the image use this syntax

```python
255 - image
```

##### To flip the image use the below syntax

```python
image[::,::-1]
```

### Project 4 - Data Normalization

```python
data = np.random.normal(size=(5,5))
data
```

```python
data_max = data.max()
data_min = data.min()
data_min,data_max
```

```python
(data - data_min)/(data_max+data_min)
```

```python
data_std = data.std()
data_mean = data.mean()
```

```python
(data - data_mean)/data_std
```
