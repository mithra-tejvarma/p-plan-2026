#### Indexing, Selection and Filtering

In this we use superstore data set to understand how it is selected how the index will work and how we can filter the data

```python
import pandas as pd
import numpy as np
```

```python
df_store = pd.read_csv('../datasets/superstore_excel.csv')
df_store
```

Now instead of just going with syntax we can ask a questions and answer to them with learning

Indexing - In pandas objects are label-based, not just position-based.


we can use normal [] brackets with df object

```python
df_store[4:10]
```

we can give the row index or column name to select data

```python
df_store['Order ID']
```

We can also select the column using membership operator with the column names having no space

```python
df_store.Region
```

We can select multiple columns by passing a list into the data frame

```python
df_store[['Region','Category']]
```

We can slice the dataframe using `:` . It is *exclusive*

```python
df_store[1:6]
```

We have now selection by Label
1. `df.loc[]` - uses label means the index value 
2. `df.at[]` - can fetch only one value 

`df.loc[]` we can use this different types
1. we can select single or subset of rows using label - `df.loc[:]` or `df.loc[val]` - inclusive end value also
2. we can select column or subset of column using label - `df.loc[:,val]` or `df.loc[:,val]`- list of columns
3. we can give both row and column indexes at once so it give matching rows and columns - `df.loc[val1,val2]`

```python
df_store.loc[2:78]
```

```python
df_store.loc[5]
```

```python
df_store.loc[:,'Profit']
```

```python
df_store.loc[:,['Profit','Discount']]
```

```python
df_store.loc[1250:2500,['Profit','Region','Discount','Category']]
```

```python
df_store.at[6,'Profit']
```

Now we can see also selection by using position
we have two methods in this 
1. `df.iloc[]` - uses position no matter what index is

2. `df.iat[]` - give fast acces to one valuse that is fetched using position

`df.iloc[]` has different usage same as loc but using position - Exclusive the end value
1. `df.iloc[where]` - select in row
2. `df.iloc[:,where]` - select in column
3. `df.iloc[where_i,where_i]` - selects both row and column

```python
df_store.iloc[4:7]
```

```python
df_store.iloc[:,12]
```

```python
df_store.iloc[1250:3750,6:14]
```

`df.iat[val1,val2]` - select a value of position row and column

```python
df_store.iat[6,12]
```

Conditions we can use them to filter the dataset
we can use all boolean conditions for them

```python
df_store[df_store.Profit <= 10]
```

```python
df_store[df_store.Sales > 1000]
```

```python
df_store[(df_store.Region=="South") & (df_store.Profit <= 0)]
```

```python
df_store[(df_store.Category=="Furniture") & (df_store.Profit <= 0)]
```

```python
df_store[df_store.Profit <= 0].loc[:,['Category','Sub-Category','Profit']]
```
