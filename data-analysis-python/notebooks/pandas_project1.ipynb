```python
leagues = ['E0','E2','E3']
frames = []

for league in leagues:
   for season in range(15,21):
        df = pd.read_csv(root+f'{season}{season+1}'+'/'+league+'.csv')
        df.insert(1,'season',season)
        frames.append(df)
```

```python
dict_seasons = {
    'England Premier League': 'E0',
    'England Championship': 'E1',
    'England League One': 'E2',
    'England League Two': 'E3',
    'England Conference': 'EC'
}
dict_dataFrames = {}

for league,code in dict_seasons.items():
    temp_frames = []
    for season in range(15,22):
        df = pd.read_csv(root+f'{season}{season+1}'+'/'+code+'.csv',encoding='unicode_escape')
        df.insert(1,'season',season)
        temp_frames.append(df)
    dict_dataFrames[league] = pd.concat(temp_frames,ignore_index=True)
```
