##### To sort an np-array we can use the np-array function *i.e* `arr.sort()`

```python
arr_rand_int2 = np.random.randint(16,size=(4,4))
arr_rand_int2
```

##### As this `arr_rand_int2` is not **1-D** array but it does sort the array for only one row means all elements in one row is sorted in the same row 

```python
arr_rand_int2.sort()
arr_rand_int2
```

##### Here we can also use parameter `arr.sort(axis)` where *axis = 0 or 1* 
1. axis = 0 represent to sort the values according to column wise means it sort all the column values
2. axis = 1 represent to sort the values according to row wise means it sort all the values in row

`arr.sort()` corresponds to `arr.sort(axis=1)`

```python
arr_rand_int2.sort(axis=0)
arr_rand_int2
```

##### To flatten an n-dim or any-shape np-array we use `arr.flatten()` it converts into 1-dim np.array and returns an 1-dim np.array

```python
arr_rand_int2.flatten()
```

##### To **Transpose** the np-arr i.e *rows* becoming *columns* and vice-verca we use `arr.T`

```python
arr_rand_int2,arr_rand_int2.T
```

##### Reshaping an array using `arr.reshape(x,y)` that reshapes an array into x,y shape
Make sure to have **x*y** elements in the array to reshape correctly without any error

```python
arr_rand_int2.reshape(2,8,order='F')
```

##### Here we have two orders they are `'C'` and `'F'` 
`C` makes to reshape on the basis of row wise filling elements

`F` makes to reshape on the basis of column wise filling elements

##### Here `arr_rand_int2` is of shape **(4,4)** but after reshape you can see the output in shape of **(2,8)**

##### Now we get another np-array function that is `arr.resize((x,y))` this make an array to rershape to any size and it automatically fills the values to **0** that are not initialised

### Indexing & Slicing

##### Indexing with `[]` to get elements of the np-array

##### `arr[1]` gives element at position 1 respecting to shape of the np-arr

```python
arr_rand_int2[1]
```

##### `arr[1,2]` gives element of position `arr[1][2]` 

```python
arr_rand_int2[1,2]
```

##### We can assign the values using `[]` like `arr[1] = 4 `

```python
arr_rand_int2[1] = 4
arr_rand_int2
```

##### Here something called broadcasting is applied in the above code. Broadcasting means some times the given number like above we give `4` value for `arr_rand_int2[1]` but it has an entire row so numpy directly make `4` as `[4,4,4,4]` matching the shape of the given np-array

##### We can also assign value to specific index like `arr[1,2] = 3`

```python
arr_rand_int2[1,2] = 3
arr_rand_int2
```

##### We can also slice the np-array as python list like `arr[1:3]` this returns the elements at index *1*,*2* 

```python
arr_rand_int2[1:3]
```

##### For multi dimension we can also specify element value to get like `arr[1:3,2]` this returns the elements at index *2* of rows *1*, *2*

```python
arr_rand_int2[1:3,2]
```

##### `arr[:2]` gives all indexes before index 2 i.e *0*, *1*

```python
arr_rand_int2[:2]
```

##### `arr[:,2]` gives all elements at index 2 from every row

```python
arr_rand_int2[:,2]
```

##### Vector Maths

4. `np.divide(arr1,arr2)` - division of two np-arrays
5. `np.power(arr1,arr2)` - power of each element of arr1 with respect to arr2
6. `np.array_equal(arr1,arr2)` - It returns the true if the both arrays have same elements
