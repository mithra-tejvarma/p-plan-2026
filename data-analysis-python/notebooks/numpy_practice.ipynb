# Importing numpy as np


```python
import numpy as np
```

### creating numpy array


```python
arr_one = np.array([1,2,3,4,5,6])
arr_one
```

### creating two dimension array


```python
arr_two= np.array([[1,2,3,4],[5,6,7,8]])
arr_two
```

### creating numpy arrays with all zeros filled using `np.zeros`

```python
arr_zero = np.zeros(5)
arr_zero
```

### creating numpy arrays with all ones filled using `np.ones`


```python
arr_ones = np.ones(5)
arr_ones
```

#### we can use `np.zeros` and `np.ones` functions by passing shape of the array needed

```python
arr_zeros_1 = np.zeros((3,4))
arr_ones_1 = np.ones((3,4))
arr_zeros_1, arr_ones_1
```

##### here these `np.zeros` and `np.ones` gives arrays in np.float data type
##### you can check the data type using `arr.dtype` where arr is the `np.array` or np object

```python
arr_zeros_1.dtype,arr_ones_1.dtype
```

##### now let us look for another function same like `zeros`,`ones` that is `np.eye`

it gives an identity matrix that is it will give an n x n matrix with 1 on diagonals and 0 on other places


```python
arr_eye = np.eye(5)
arr_eye
```

##### another function to create np-array is using `np.linspace` this creates the array from start to stop on giving divison length between them

```python
arr_linspace = np.linspace(3,30,10)
arr_linspace
```

##### here the array dtype is `np.float64`

```python
arr_linspace.dtype
```

##### Now another function to create numpy-array is `np.arange` it creates a range of values with a *start* , *stop* , *step* in it

```python
arr_arange = np.arange(16,32)
arr_arange
```

##### the `np.arange` array shape is always 1-dim.

```python
arr_arange.shape
```

##### another function for creating np-arrays is `np.full` it can take the shape that needed to create and a value to fill the entire array with that

```python
arr_full = np.full((3,4),5)
arr_full
```

##### another type to create a np-array is to use `np.random.rand`  
1. we can use it two type if we use `np.random.rand` directly it gives valuse in range between 0 and 1
2. if we use `np.random.rand()*100` it gives float between 0 and 100 

```python
arr_rand = np.random.rand(3,4)
arr_rand
```

```python
arr_rand_1 = np.random.rand(3,4)*100
arr_rand_1
```

##### the last function is `np.random.randint` which takes an integer number and a tuple of shape and it returns a random integer between given starting and stoping integer


```python
arr_rand_int = np.random.randint(10,size=(3,4))
arr_rand_int
```
